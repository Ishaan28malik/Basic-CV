```python
import Tkinter as tk
#import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

def app():
    root= tk.Tk()
    root.config(background='white')
    #root.geometry("1000x700")

    Data1 = {'Country': ['US','IND','GER','UK','FR','RS'],
            'GDP_Per_Capita': [45000,42000,52000,49000,47000,56000]
           }

    df1 = DataFrame(Data1, columns= ['Country', 'GDP_Per_Capita'])
    df1 = df1[['Country','GDP_Per_Capita']].groupby('Country').sum()



    Data2 = {'Year': [1990,1995,2000,2005,2010,2015],
            'Unemployment_Rate': [9.8,6.9,7,6.5,6.2,5.5]
           }

    df2 = DataFrame(Data2,columns=['Year','Unemployment_Rate'])
    df2 = df2[['Year','Unemployment_Rate']].groupby('Year').sum()



    Data3 = {'States_of_India': ['Ker','Pun','Goa','Mum','Guj','Raj','Del','Har'],
             'Avg_Air_Pollution_Level': [112,196,86,201,145,162,267,211]
           }

    df3 = DataFrame(Data3,columns=['States_of_India','Avg_Air_Pollution_Level'])
    #df3 = df3[['States_of_India','Avg_Air_Pollution_Level']].groupby('States').sum()
    
    
    figure1 = plt.Figure(figsize=(6,5), dpi=80)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, root)
    bar1.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    df1.plot(kind='bar', legend=True, ax=ax1)
    ax1.set_title('Country Vs  GDP Per Capita')

    
    figure2 = plt.Figure(figsize=(6,5), dpi=80)
    ax2 = figure2.add_subplot(111)
    line2 = FigureCanvasTkAgg(figure2, root)
    line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    df2.plot(kind='line', legend=True, ax=ax2, color='r',marker='o', fontsize=10)
    ax2.set_title('Year Vs  Unemployment Rate')

    
    figure3 = plt.Figure(figsize=(6,5), dpi=80)
    ax3 = figure3.add_subplot(111)
    ax3.scatter(df3['States_of_India'],df3['Avg_Air_Pollution_Level'], color = 'g')
    scatter3 = FigureCanvasTkAgg(figure3, root) 
    scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    #df3.scatter(kind='scatter', legend=True, ax=ax3)
    ax3.legend() 
    ax3.set_xlabel('States')
    #ax3.set_ylabel('Stock Index Price')
    ax3.set_title('States Vs  Air Pollutants')

    root.mainloop()
if __name__ == '__main__':
    app()    
```
